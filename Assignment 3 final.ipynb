```python
import epitran
import pandas as pd
```

```python
english_words_list = ['Apple','apricot','pear','strawberry','raspberry','cherry','peach','carrot','cauliflower','pea']
hindi_words_list = ['सेब','खुबानी','नाशपाती','स्ट्रॉबेरी','रसभरी','चेरी','आड़ू','गाजर','गोभी','मटर']
```

```python
ipa_hindi_list = []
ipa_index = []
count = 1
epH = epitran.Epitran('hin-Deva')
for h in hindi_words_list:
    ipa_hindi_list.append(epH.transliterate(h))
    ipa_index.append(count)
    count = count+1
```

```python
data = {'English':english_words_list,
        'Hindi':hindi_words_list,
        'Hindi to IPA':ipa_hindi_list
       }
df_german_to_ipa = pd.DataFrame(data, index=ipa_index)
df_german_to_ipa
```

```python
ipa_dict_1 = {
    '̤':'ھ',
    'a':'ا',
    'ʰ':'ھ',
    'p':'پ',
    'b':'ب',
    'pʰ':'پھ',
    'bʰ':'بھ',
    'm':'م',
    'mʰ':'مھ',
    't':'ت',
    't̪':'ت',
    't̪ʰ':'تھ',
    'd̪':'د',
    'd̪ʰ':'دھ',
    'n':'ن',
    'nʰ':'نھ',
    'Ŋ':'نگ',
    'ʈ':'ٹ',
    'ɖ':'ڈ',
    'ʈʰ':'ٹھ',
    'ɖʰ':'ڈھ',
    'k':'ک',
    'ɡ':'گ',
    'kʰ':'کھ',
    'gʰ':'گھ',
    'q':'ق',
    'ʔ':'ع',
    'f':'ف',
    'v':'و',
    's':'س',
    'z':'ز',
    'ʃ':'ش',
    'ʒ':'ژ',
    'ɣ':'غ',
    'x':'خ',
    'h':'ح',
    'ʧ':'چ',
    'ʧʰ':'چھ',
    'ʤ':'ج',
    'ʤʰ':'جھ',
    'r':'ر',
    'rʰ':'رھ',
    'ɽ':'ڑ',
    'ɽʰ':'ڑھ',
    'j':'ی',
    'l':'ل',
    'lʰ':'لھ',
    'vʰ':'وھ',
    'jʰ':'یھ',
    'i':'ی',
    'e':'ی',
    'æ':'ـِىـِ',
    'u':'و ُ',
    'o':'و',
    'ɔ':'و َ',
    'ɑ':'آ',
    'ɪ':'َ'
}
ipa_dict_1
```

```python
def split(word):
    return [char for char in word]
```

```python
ipa_hindi_after = []
for hb in ipa_hindi_list:
    a= split(hb)
    test_string = ''
    for n, i in enumerate(a):    
        if (i == 'ː'):
            a[n] = ''
        if(i == '͡'):
            a[n] = '' 
        test_string = test_string+a[n]
    ipa_hindi_after.append(test_string)
```

```python
a = 0
ipa_to_urdu = []
for z in ipa_hindi_after:
    word = split(z)
    result = ''
    for n,x in enumerate(word):
    #     print(x)
        for k,v in ipa_dict_1.items():
            if n==0 and x=='a':
                result = 'آ'            
            elif x == k:
                result = result + v
```

```python
data = {'English':english_words_list,
        'Hindi':hindi_words_list,
        'Hindi to IPA':ipa_hindi_list,
        'Urdu':ipa_to_urdu
       }
df_german_to_ipa = pd.DataFrame(data, index=ipa_index)
df_german_to_ipa
```
